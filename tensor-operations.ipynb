```python
import torch

# 1D Tensor
tensor_1d = torch.tensor([1, 2, 3])
# 2D Tensor
tensor_2d = torch.tensor([[1, 2], [3, 4]])
# 3D Tensor
tensor_3d = torch.tensor([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])

print(tensor_2d)
```

```python
import numpy as np

# 1D Array
array_1d = np.array([1, 2, 3])
# 2D Array
array_2d = np.array([[1, 2], [3, 4]])
# 3D Array
array_3d = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])

print(array_2d)
```

```python
a = torch.tensor([1, 2, 3])
b = torch.tensor([4, 5, 6])

add = a + b
print(add)
sub = a - b
print(sub)
mul = a * b
print(mul)
div = a / b
print(div)
```

```python
a_np = np.array([1, 2, 3])
b_np = np.array([4, 5, 6])

# NumPy
add_np = a_np + b_np
print(add_np)
sub_np = a_np - b_np
print(sub_np)
mul_np = a_np * b_np
print(mul_np)
div_np = a_np / b_np
print(div_np)
```

```python
# Dot product (1D)
dot_pt = torch.dot(a, b)
dot_np = np.dot(a_np, b_np)
print(dot_pt)
print(dot_np)

# Matrix Multiplication (2D)
mat1 = torch.tensor([[1, 2], [3, 4]])
mat2 = torch.tensor([[5, 6], [7, 8]])
matmul_pt = torch.matmul(mat1, mat2)

mat1_np = np.array([[1, 2], [3, 4]])
mat2_np = np.array([[5, 6], [7, 8]])
matmul_np = np.matmul(mat1_np, mat2_np)

print(matmul_pt)
print(matmul_np)
```

```python
x = torch.tensor([[10, 20, 30], [40, 50, 60]])

# Extract row
row = x[1]
print(row)
# Extract column
col = x[:, 1]
print(col)
# Boolean Masking
mask = x > 25
masked = x[mask]
print(masked)
```

```python
x_np = np.array([[10, 20, 30], [40, 50, 60]])
row_np = x_np[1]
print(row_np)
col_np = x_np[:, 1]
print(col_np)
masked_np = x_np[x_np > 25]
print(masked_np)
```

```python
# PyTorch
a = torch.tensor([[1], [2], [3]])  # shape (3,1)
b = torch.tensor([4, 5, 6])        # shape (3,)
result = a + b  
print(result.size())

# NumPy
a_np = np.array([[1], [2], [3]])
b_np = np.array([4, 5, 6])
result_np = a_np + b_np
```

```python
x = torch.tensor([1.0, 2.0, 3.0])

# Out-of-place (returns a new tensor)
y = x + 1
print(y)

# In-place (modifies the original tensor)
x.add_(1)  
print(x)
```
